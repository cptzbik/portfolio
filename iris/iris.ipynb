# Iris EDA Data Analysis

This notebook presents an exploratory data analysis (EDA) of the famous Iris flower dataset. It includes a description of the dataset, characteristics of the three Iris species, and an overview of the features (sepal length and width, petal length and width).

## Dataset Overview

The Iris dataset contains measurements for three species of Iris flowers: setosa, versicolor, and virginica. The features include sepal length, sepal width, petal length, and petal width.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('25__iris.csv', sep=",")
df = df
```

### Data Completeness

The dataset is complete and contains no missing values. All features are available for each sample.

```python
df.info()
```

### Class Distribution

The dataset contains an equal number of samples for each Iris species, which allows for balanced analysis and modeling.

### Correlation Analysis

The correlation matrix below shows the relationships between the features. Strong correlations can help in feature selection and understanding the data structure.

### Feature Distributions

The following plots show the distributions of each feature for the three Iris species. These visualizations help identify differences and similarities between the species.

```python
df.isnull()
```

### Outlier Detection

Outliers are identified and visualized to better understand the data and ensure robust modeling.

### Summary of Observations

The analysis highlights the main differences between the Iris species and provides insights for further modeling and classification tasks.

```python
df_columns = df.iloc[:, :-1]
corr = df_columns.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

### Visualization: Pair Plots

Pair plots allow for a quick visual comparison of feature relationships and class separability.

```python
df.plot(kind='scatter', x="długość kielicha (sepal length)", y='długość płatka (petal length)')
df.plot(kind='scatter', x="szerokość kielicha (sepal width)", y='szerokość płatka (petal width)')
xlabel = 'szerokość kielicha i długość'
```

```python
for class_name, group in df.groupby('klasa (class)'):
    plt.scatter(group['długość kielicha (sepal length)'], group['długość płatka (petal length)'], label=class_name)
plt.xlabel('długość kielicha (sepal length)')
plt.ylabel('długość płatka (petal length)')
plt.legend(title='Klasy obiektów:')
plt.title('Długość kielicha (sepal length) vs długość płatka (petal length)')
plt.show()
```

### Visualization: Boxplots

Boxplots show the spread and outliers for each feature, grouped by Iris species.

### Visualization: Histograms

Histograms provide a detailed look at the distribution of each feature for all classes.

```python
plt.figure(figsize=(12, 10))
sns.boxplot(data=df)
plt.title('Boxplot dla całego zbioru danych iris')
plt.xticks(rotation=90)
plt.show()
```

### Visualization: Scatter Plots

Scatter plots help visualize the relationships between pairs of features and the separation between classes.

### Final Remarks

This EDA provides a foundation for further machine learning and classification tasks using the Iris dataset.

```python
setosa = df[df['klasa (class)'] == 'Iris-setosa']
versicolor = df[df['klasa (class)'] == 'Iris-versicolor']
virginica = df[df['klasa (class)'] == 'Iris-virginica']

plt.figure(figsize=(10,6))
plt.hist(setosa['długość płatka (petal length)'], bins=20, alpha=0.4, label='Iris-Setosa')
plt.hist(versicolor['długość płatka (petal length)'], bins=20, alpha=0.4, label='Iris-Versicolor')
plt.hist(virginica['długość płatka (petal length)'], bins=20, alpha=0.4, label='Iris-virginica')
plt.xlabel('Długość płatka (cm):')
plt.ylabel('Liczba wystąpień:')
plt.title('Histogram długości płatka dla trzech klas irysów.')
plt.legend(loc='upper center')
```
